## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd

# Dataframe:
df = pd.read_csv('../weather-raw.csv')
display(df.head(5))
```

```python
# La tabla tiene muchos valores vacíos, por lo que evaluar la temperatura a diario no tiene mucho sentido.
# Voy a calcular estadísticos descriptivos por meses.

# Creo un dataframe vacío para rellenarlo luego:
# Relleno los meses extrayéndolos de df con un set:
df2 = pd.DataFrame({'month': list(set(df['month'])), 'tmin':[0]*len(list(set(df['month']))), 'tmax':[0]*len(list(set(df['month'])))})
display(df2)
```

```python
# min of each month:
dftmin = df[df['element'] == 'tmin']
df2['tmin'] = list(dftmin.iloc[:,4:35].min(skipna=True, axis=1))
# df2['tmin'] = list(df.iloc[1::2,4:35].min(skipna=True, axis=1)) # se puede hacer lo mismo sin crear una tabla intermedia, cogiendo los valores de 2 en 2

# max of each month:
dftmax = df[df['element'] == 'tmax']
df2['tmax'] = list(dftmax.iloc[:,4:35].max(skipna=True, axis=1))
# df2['tmax'] = list(df.iloc[::2,4:35].max(skipna=True, axis=1)) # se puede hacer lo mismo sin crear una tabla intermedia, cogiendo los valores de 2 en 2

display(df2)
```

```python
import calendar
df2['month'] = df2['month'].apply(lambda x: calendar.month_name[x])
display(df2)
```

```python
# También se podría conseguir usando .melt:
melted_cols = df.columns[4:-1]
DF_melted = df.melt(id_vars=["month",'element'],value_vars=melted_cols)
display(DF_melted.head(5))
```

```python
TMAX = DF_melted[DF_melted['element']=='tmax']
TMIN = DF_melted[DF_melted['element']=='tmin']

tmaxlist = []
for e in set(df['month']):
    aux = TMAX[TMAX['month']==e]
    tmaxlist.append(aux['value'].max(skipna=True, axis=0))

tminlist = []
for e in set(df['month']):
    aux = TMIN[TMIN['month']==e]
    tminlist.append(aux['value'].min(skipna=True, axis=0))
    
print(tmaxlist)
print(tminlist)
```
